```python
for n in range(1,10):
    print(3, 'x', n, '=', 3*n)
```

```python
def gugudan(x):
    for n in range(1,10):
        print(x, 'X', n, '=', x*n)
```

```python
gugudan(5)
```

```python
gugudan('오마이걸')
```

```python
def gugudan_new(x):
    if x in list(range(1,10)):
        print(x, '단입니다', '\n',sep='')
        for n in range(1,10):
            print(x, 'X', n, '=', x*n)
    else:
        print('틀려버렸지 뭐야!')
```

```python
gugudan_new(7)
```

```python
gugudan_new('오마이걸')
```
